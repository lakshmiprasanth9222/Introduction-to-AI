```python
#pre oversampling
from imblearn.over_sampling import RandomOverSampler
from collections import Counter


X = df.drop(columns='diagnosis')  
y = df['diagnosis']               

print("Pre Oversampling:", Counter(y))
```

```python
#Assigning target column as diagnosis
x=df.drop(columns=['diagnosis'])
y=df['diagnosis']
```

```python
#Importing train-test-split
from sklearn.model_selection import train_test_split
```

```python
#splitting data for training and testing
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python

print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
#training the model
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(xtrain,ytrain)
```

```python
#saving the model
import joblib
joblib.dump(log,'logrm_model.joblib')
```

```python
#loading the model
logrm_model=joblib.load('logrm_model.joblib')
```

```python
#Importing libraries for performance testing
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix,roc_auc_score,roc_curve
```

```python
#testing the model
y_predict=logrm_model.predict(xtest)

logrm_accuracy=accuracy_score(ytest,y_predict)
logrm_conf_matrix=confusion_matrix(ytest,y_predict)
logrm_class_report= classification_report(ytest,y_predict)
```

```python
#print the metrics
print(f'Accuracy: {logrm_accuracy: .2f}')
print(f'\nConfusion Matrix:\n{logrm_conf_matrix}')
print(f'\nClassification Report:\n{logrm_class_report}')

#confusion matrix plotting
plt.figure(figsize=(5,4))
sns.heatmap(logrm_conf_matrix, annot=True,fmt="d",cmap='Reds')
plt.title('Logistic Regression')
plt.ylabel('Actual Values ')
plt.xlabel('Predicted Values ')
plt.show()
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from scipy.stats import uniform

#pipeline with scaling and logistic regression
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('logreg', LogisticRegression(solver='liblinear'))  # 'liblinear' supports L1 and L2
])

# hyperparameter space
param_dist = {
    'logreg__C': uniform(loc=0.01, scale=10),  # Regularization strength
    'logreg__penalty': ['l1', 'l2'],           # Type of regularization
    'logreg__max_iter': [100, 200, 300]        # Iterations for convergence
}

#RandomizedSearchCV
random_search = RandomizedSearchCV(
    pipeline,
    param_distributions=param_dist,
    n_iter=20,
    scoring='accuracy',
    cv=5,
    verbose=1,
    random_state=42,
    n_jobs=-1
)

# Fit to training data
random_search.fit(xtrain, ytrain)

# Evaluating the model
best_model = random_search.best_estimator_
y_pred = best_model.predict(xtest)
postlogrm_conf_matrix=confusion_matrix(ytest,y_pred)
print(postlogrm_conf_matrix)

print(" Best Hyperparameters:", random_search.best_params_)
print("\n Accuracy on test data:", accuracy_score(ytest, y_pred))
print("\n Classification Report:\n", classification_report(ytest, y_pred))
```
